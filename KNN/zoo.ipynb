Standarized the variables

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(zoo_df.drop('type',axis=1))
scale_feature=scaler.transform(zoo_df.drop("type",axis=1))
scale_feature
```

```python
#Taking the independent feature
zoodf=pd.DataFrame(scale_feature,columns=zoo_df.columns[:-1])  
zoodf.head(5)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(scale_feature,zoo_df['type'],
                                              test_size=0.30)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
pred
```

# Prediction and Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

# Choosing K value

```python
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i=knn.predict(X_test)
    error_rate.append(np.mean(pred_i !=y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='black',linestyle='dashed',marker='o')
plt.title('Erro Rate vs. K value')
plt.xlable('K')
plt.ylabel('Error Rate')
           
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
