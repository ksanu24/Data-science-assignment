```python
df=data.pivot_table(values='rating', index="userid", columns="tittle",fill_value=0)
df
```

###  Building the Recommendation System using Cosine simil

```python
similar_users=1-pairwise_distances(df.values,metric='cosine')
similar_users
```

```python
sim_df=pd.DataFrame(similar_users)
sim_df
```

```python
sim_df.index=df.index
sim_df.columns=df.index
sim_df
```

```python
np.fill_diagonal(similar_users,0)
sim_df
```

```python
sim_df=sim_df.replace(0,np.nan)
sim_df
```

```python
similarity=pd.DataFrame(sim_df.idxmax(axis=1,skipna=True))
similarity
```

### Finding the similarity of the book based on the rating

```python
# Function to find simlarity with other books based on ratings
def book_similarity(data, book_title):
    rating = data[book_title]
    similar = data.corrwith(rating)
    corr_df = pd.DataFrame(similar, columns=['Correlation'])
    corr_df.dropna(inplace=True)
    print(f'---Top 5 similar books for book "{book_title}" are:')
    print(corr_df[(corr_df['Correlation'] > 0) & (corr_df.index != book_title)].sort_values(by='Correlation', ascending=False).head(5))
    print("____________________\n")
```

```python
data['tittle'].value_counts().head(10)
```

```python
famous_books=data['tittle'].value_counts().head(10).keys()
```

#### Top 5 books similar to top 10 famous books

```python
for books in famous_books:
    book_similarity(df,books)
```
