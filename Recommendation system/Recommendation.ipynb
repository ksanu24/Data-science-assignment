```python
book=pd.read_csv("book.csv")
book
```

```python
#rename the columns
book.columns=['sn','user','tittle','rating']
```

```python
book
```

```python
books=book.drop(['sn'],axis=1)
```

```python
books
```

```python
#number of unique users in the dataset
len(books.user.unique()) 
```

```python
bookdf=books.dropna()
bookdf
```

```python
book_df=books.reset_index(drop=True).pivot_table(index='user',
                 columns='tittle',
                 values='rating')
```

```python
book_df
```

```python
book_df.index = books.user.unique()
```

```python
#Impute those NaNs with 0 values
book_df.fillna(0, inplace=True)
```

```python
book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book2 = 1 - pairwise_distances( book_df.values,metric='cosine')#distance using cosine 
```

```python
book2
```

```python
#Store the results in a dataframe
book2_df = pd.DataFrame(book2)
```

```python
book2_df
```

```python
#Set the index and column names to user ids 
book2_df.index = bookdf.user.unique()
book2_df.columns = bookdf.user.unique()
```

```python
book2_df.iloc[0:9999, 0:]
```
