```python
## Correlation Matrix
sns.set(style="white")

# Compute the correlation matrix
corr = bank.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(10, 10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
banks_df=pd.concat([outcome,months,contacts,house,loans,defaults,educations,maritals,jobs],axis=1) #added dummy data 
banks_df
```

```python
banks_df.columns
```

```python
# Normalizes
df1= (banks_df - np.min(banks_df)) / (np.max(banks_df) - np.min(banks_df)).values
```

```python
from sklearn.model_selection import train_test_split
x=df1.iloc[:,0:42]
y=bank['y']
```

We will split our data. 80% of our data will be train data and 20% of it will be test data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Logistic regression and fit the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train) #we will fit the tarin data in model
```

```python
classifier.coef_
```

```python
#Predict for X dataset
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
from sklearn import metrics
print(metrics.accuracy_score(y_pred,y_test))
confusion_matrix = confusion_matrix(y_pred,y_test)
print (confusion_matrix)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(metrics.classification_report(y_pred,y_test))
```
